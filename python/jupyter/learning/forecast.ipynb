```r
install.packages('forecast', dependencies = TRUE)
```

```r
library("forecast")
```

```r
fraser <- scan("fraser.txt")
```

```r
plot(fraser)
```

```r
fraser.ts <- ts(fraser, frequency=12, start=c(1913,3))
```

```r
fraser.stl = stl(fraser.ts, s.window="periodic")
```

```r
monthplot(fraser.stl)
```

```r
seasonplot(fraser.ts)
```
