```python
x = np.linspace(0,10,1000)
y = np.cumsum(np.random.randn(1000,6), 0)
```

```python
plt.plot(x, y)
plt.show()
```

```python
data = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])

# 'normed' kwarg is deprecated, use 'density' instead
for dataset in 'ab':
    plt.hist(data[dataset], density=True, alpha=0.5)
    
plt.show()
```

```python
for dataset in 'ab':
    sns.kdeplot(data[dataset], shade=True)
    
plt.show()
```
